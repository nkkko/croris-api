```python
import pickle

# Load the pickle file
data = []
with open('/Users/nikola/Downloads/publications.pkl', 'rb') as file:
    while True:
        try:
            batch = pickle.load(file)
            data.extend(batch)
        except EOFError:
            break

# Check the type of the loaded data
print(type(data))

# Preview the first few elements and print its size
print(data[:5])
print("Size of the list:", len(data))
```
